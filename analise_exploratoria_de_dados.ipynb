### **Instalando e descompactando arquivos .rar**

Como há arquivos comprimidos em formato `.rar`, precisamos instalar a ferramenta `unrar` para descompactá-los antes de carregar os dados.

```python
!apt-get install unrar
```

```python
# Encontra todos os arquivos .rar
rar_files = glob.glob('**/*.rar', recursive=True)

print(f"Found {len(rar_files)} .rar files to extract.")

# Descompacta cada arquivo .rar
for rar_file in rar_files:
    try:
        # O comando 'unrar x' extrai os arquivos com seus nomes de diretório completos
        # O destino da extração será o diretório pai do arquivo .rar
        output_dir = os.path.dirname(rar_file)
        if not output_dir:
            output_dir = '.' # Se não houver diretório pai, extrai no diretório atual
        print(f"Extracting {rar_file} to {output_dir}")
        !unrar x -o- '{rar_file}' '{output_dir}'
        print(f"Successfully extracted {rar_file}")
    except Exception as e:
        print(f"Error extracting {rar_file}: {e}")
```

### **Carregando e concatenando todos os dados CSV**

Agora que todos os arquivos (benignos, mirai e gafgyt) estão descompactados e disponíveis como `.csv`, vamos carregá-los em um único DataFrame.

```python
all_files = glob.glob('**/*.csv', recursive=True)

# Remove arquivos de metadados ou estrutura se houver algum no caminho
data_files = [
    f for f in all_files
    if 'demonstrate_structure' not in f and 'description' not in f
]

# Carrega todos os arquivos em uma lista de DataFrames
df_list = [
    pd.read_csv(filepath)
    for filepath in data_files
]

#Concatena todos os DataFrames em um único objeto
df=pd.concat(df_list, ignore_index=True)

#Limpeza dos Nomes das Colunas
df.columns = df.columns.str.strip()
```

####**Danmini doorbell**

```python
import pandas as pd
import glob
import os

# 1. Definição do Dispositivo
DEVICE_NAME = 'Danmini_Doorbell'

# 2. Busca de arquivos: Encontra todos os CSVs dentro da pasta
search_path = f'{DEVICE_NAME}/*.csv'
device_files = glob.glob(search_path)

print(f"Encontrados {len(device_files)} arquivos CSV para o dispositivo: {DEVICE_NAME}")

df_list = []

for filepath in device_files:
    # 3. Extração do Rótulo de Ataque
    filename = os.path.basename(filepath)
    attack_type = filename.split('_')[0]

    print(f"  -> Processando: {filename} (Tipo: {attack_type})")

    try:
        # 4. Leitura e Rotulação
        # Usando low_memory=False para carregar grandes arquivos de forma mais robusta
        df_aux = pd.read_csv(filepath, low_memory=False)

        df_aux['Device'] = DEVICE_NAME
        df_aux['Attack_Type'] = attack_type

        df_list.append(df_aux)

    except Exception as e:
        print(f"  -> ERRO ao ler {filepath}: {e}")

# 5. Concatenação e Limpeza
if df_list:
    df_device = pd.concat(df_list, ignore_index=True)

    # Limpeza dos Nomes das Colunas
    df_device.columns = df_device.columns.str.strip()

    print("\n--- RESUMO DO DATAFRAME ---")
    print(f"DataFrame final de {DEVICE_NAME} criado com {len(df_device):,} linhas.")
    print(f"Verificação de Rótulos (Attack_Type):")
    print(df_device['Attack_Type'].value_counts())
else:
    print("Nenhum DataFrame foi carregado. Verifique os caminhos e a descompactação.")
```
